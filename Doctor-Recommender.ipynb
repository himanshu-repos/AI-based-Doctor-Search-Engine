```python
import wikipediaapi

# Specify a user agent to comply with Wikipedia's policy
user_agent = "YourProjectName/1.0 (https://yourwebsite.com; your-email@example.com) Wikipedia-API/0.5"
wiki_wiki = wikipediaapi.Wikipedia('en', headers={"User-Agent": user_agent})

# Function to get symptoms for a given disease from Wikipedia
def get_symptoms_enhanced(disease_name):
    """
    Fetches 'Signs and symptoms' section from a Wikipedia page for a given disease.
    If 'Signs and symptoms' is not found, it attempts to fetch any relevant information.
    """
    try:
        # Fetch the Wikipedia page
        page = wiki_wiki.page(disease_name)
        
        # If page does not exist, return message
        if not page.exists():
            return "Wikipedia page not found."
        
        # Look for 'Signs and symptoms' section
        symptoms = ""
        for section in page.sections:
            if 'symptom' in section.title.lower():
                symptoms = section.text
                break
        
        # Alternative section if symptoms are not directly available
        if not symptoms:
            for section in page.sections:
                if 'overview' in section.title.lower() or 'general' in section.title.lower():
                    symptoms = section.text
                    break
        
        # Return symptoms text or a message if not found
        return symptoms.strip() if symptoms else "Symptoms section not found."
    
    except Exception as e:
        return f"Error occurred: {e}"

# Add a 'Symptoms' column by applying the function to the 'Disease' column
description_df['Symptoms'] = description_df['Disease'].apply(get_symptoms_enhanced)

print("Symptoms column added to the DataFrame.")
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# now ill make a transform data function in which i will first change  all the words into lower case
# then ill tokenize all the words in the string of symptom


# then i'll  remove all the enlgish stopwords from the data
# then i'll  remove all the string punctuations from the data

# After this i'll apply stemming which will make the same looking word from the same verb  like - >  dances ,dance  , dancing , dancer
```

```python
def transform_data(text):
    text = text.lower()
    text = nltk.word_tokenize(text)

    
    y = []
    for i  in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            y.append(i)

    text = y[:]
    y.clear()

    for i in text:
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
def filter_input(text):
    text  = transform_data(text)

 # repeating the words in text 3 times  , for increasing some weight of some useful word in the user input
    words = text.split()
    tripled_words = ' '.join(words * 3)
    input_vector = tfidf.transform([tripled_words]).toarray()
    return input_vector
```

```python
def assign_location_data_with_mapping(df):
    random.seed(42) 
    pincodes_data = []
    locations_data = []
    addresses = []

    for _ in range(len(df)):
        # Randomly choose a pin code
        pincode = random.choice(list(pincode_to_location.keys()))
        #  pincodes corresponding location
        location = pincode_to_location[pincode]
        # Now ill generate random street number and house number
        street_no = random.randint(1, 50)
        house_no = random.randint(1, 100)
        address = f"House No. {house_no}, Street No. {street_no}, {location}"

        pincodes_data.append(pincode)
        locations_data.append(location)
        addresses.append(address)

    df['Pincode'] = pincodes_data
    df['Location'] = locations_data
    df['Address'] = addresses
    return df
```

```python
def get_doctor_details(doctor_name, doc_df):
    # Search for the doctor in the original dataset
    doctor_info = doc_df[doc_df['Name'] == doctor_name].iloc[0]
    return {
        'Name': 'Dr. ' +  doctor_info['Name'],
        'Phone': doctor_info['Phone Number'],
        'Timing': doctor_info['Timing'],
        'Specialist': doctor_info['Specialist'],
        'Address': doctor_info['Address'],
        'Pincode': doctor_info['Pincode']
    }
```

```python
def get_nearest_doctors(user_pincode, filtered_doctor_names, dataset, top_n=4):
    doctor_distances = []

    # Iterate over the filtered list of doctor names
    for doctor_name in filtered_doctor_names:
        # Get doctor details from the dataset
        doctor_details = get_doctor_details(doctor_name, dataset)
        
        # Calculate the distance (absolute difference between pincode values)
        distance = abs(int(doctor_details['Pincode']) - int(user_pincode))
        
        # Append the doctor details along with the distance
        doctor_distances.append((doctor_details, distance))
    
    # Sort the doctor list based on distance (ascending)
    doctor_distances.sort(key=lambda x: x[1])
    
    # Get the top N nearest doctors
    top_doctors = doctor_distances[:top_n]

    # Return the details of the top N nearest doctors
    return [doctor[0] for doctor in top_doctors]
```

```python
def recommend_doctor(inp_vector, vector, user_pincode):
    # Symptom-based filtering using cosine similarity
    similarity_scores = cosine_similarity(inp_vector, vector) 
    sorted_indices = similarity_scores.argsort()[0][::-1]  
    
    similar_disease = sorted_indices[:2]
    top_disease = description_df.iloc[similar_disease]
    
    # list of predicted diseases
    predicted_disease = []
    for disease in top_disease.itertuples():
        predicted_disease.append(disease.Disease)
    
    # Specialist list filtering
    specialist_list = []
    for idx, specialist in specialist_df['Drug Reaction'].items():  
        if specialist in predicted_disease:
            specialist_list.append(specialist_df.loc[idx, 'Allergist']) 

    # Geting the doctors who are specialists
    doctor_list = []
    for idx, doctors in doc_df['Specialist'].items():
        if doctors in specialist_list:
            doctor_list.append(doc_df.loc[idx, 'Name'])

    # Location-based filtering from the doctor_list
    doctor_distances = []

    for doctor_name in doctor_list:
        # Geting doctor details from the dataset
        doctor_details = get_doctor_details(doctor_name, doc_df)
        
        # Calculating the abslute distance
        distance = abs(int(doctor_details['Pincode']) - int(user_pincode))
        
        # Append the doctor details along with the distance
        doctor_distances.append((doctor_details, distance))
    
    # Sorting doctors by distance 
    doctor_distances.sort(key=lambda x: x[1])
    top_doctors = doctor_distances[:4]

    return [doctor[0] for doctor in top_doctors]
```

```python
from IPython.display import display, IFrame

def recommend_doctor_with_map(inp_vector, vector, user_pincode):
    # Step 1: Symptom-based filtering using cosine similarity
    similarity_scores = cosine_similarity(inp_vector, vector) 
    sorted_indices = similarity_scores.argsort()[0][::-1]  
    
    similar_disease = sorted_indices[:2]
    top_disease = description_df.iloc[similar_disease]
    
    # Create list of predicted diseases
    predicted_disease = []
    for disease in top_disease.itertuples():
        predicted_disease.append(disease.Disease)
    
    # Step 2: Specialist list filtering
    specialist_list = []
    for idx, specialist in specialist_df['Drug Reaction'].items():  
        if specialist in predicted_disease:
            specialist_list.append(specialist_df.loc[idx, 'Allergist']) 

    # Step 3: Get the doctors who are specialists
    doctor_list = []
    for idx, doctors in doc_df['Specialist'].items():
        if doctors in specialist_list:
            doctor_list.append(doc_df.loc[idx, 'Name'])

    # Step 4: Location-based filtering (based on pincode)
    doctor_distances = []

    for doctor_name in doctor_list:
        # Get doctor details from the dataset
        doctor_details = get_doctor_details(doctor_name, doc_df)
        
        # Calculate the distance (absolute difference between pincode values)
        distance = abs(int(doctor_details['Pincode']) - int(user_pincode))
        
        # Append the doctor details along with the distance
        doctor_distances.append((doctor_details, distance))
    
    # Step 5: Sort doctors by distance and return top N closest doctors
    doctor_distances.sort(key=lambda x: x[1])
    top_doctors = doctor_distances[:4]  # Adjust number of doctors as needed

    # Step 6: Plot doctors' locations on a map using Folium
    doctor_map = folium.Map(location=[26.2183, 78.1828], zoom_start=12)  # Center map on Gwalior
    
    for doctor in top_doctors:
        doctor_details = doctor[0]
        
        # Extract individual latitude and longitude values
        latitude = doctor_details['Latitude']
        longitude = doctor_details['Longitude']
        
        # Add a marker for each doctor
        folium.Marker(
            location=[latitude, longitude],  # Pass individual float values
            popup=f"Doctor: {doctor_details['Name']}<br>Specialist: {doctor_details['Specialist']}<br>Phone: {doctor_details['Phone Number']}<br>Address: {doctor_details['Address']}<br>Pincode: {doctor_details['Pincode']}",
            icon=folium.Icon(color='blue', icon='info-sign')
        ).add_to(doctor_map)

    # Save the map to an HTML file and display it in the notebook
    map_file = "doctor_locations_map.html"
    doctor_map.save(map_file)

    # Display map in the notebook
    display(IFrame(map_file, width=800, height=600))
    
    # Return doctor details as usual
    return [doctor[0] for doctor in top_doctors]

def get_doctor_details(doctor_name, doc_df):
    # Helper function to get details of a doctor from the dataset
    return doc_df[doc_df['Name'] == doctor_name].iloc[0]
```
